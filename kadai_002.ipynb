```python
import requests
from bs4 import BeautifulSoup
import re

# ① Webページ取得（例：青空文庫URL）
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"
res = requests.get(url)
res.encoding = res.apparent_encoding

# ② BeautifulSoupで本文抽出
soup = BeautifulSoup(res.text, "html.parser")
text = soup.get_text()

# ③ 前処理（HTMLタグ・不要文字除去）
text = re.sub(r'\n+', '\n', text)             # 改行を整形
text = re.sub(r'［.*?］', '', text)           # ルビ・注釈削除
text = re.sub(r'[｜　]', '', text)            # 不要記号削除
text = text.strip()

# ④ 最初の一文を抽出
first_sentence = text.split("。")[0] + "。"
print("最初の文:", first_sentence)

# ⑤ 手動で品詞ごとに区切り＋ストップワード除去例
sentence = "近頃 は 文壇 の 状態 に 就い て 、 私 は 少し 気 に なっ て いる こと が ある と 思っ て い ます 。"
words = sentence.split()

# ストップワード例（助詞や一般的機能語）
stopwords = {"は", "に", "の", "が", "を", "と", "て", "で", "する", "いる", "ある", "です", "ます"}

cleaned = [w for w in words if w not in stopwords]

print("形態素分解（手動）:", words)
print("ストップワード除去後:", cleaned)
```
